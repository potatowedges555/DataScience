### 1. [簡答題] 請問下列四種不同的 DataFrame 選取結果有什麼差異？

```
df.loc[ '2013-01-01', 'A'] 
df.loc[ '2013-01-01', ['A', 'B'] ] 
df.loc[ '2013-01-01':'2013-01-02', 'A' ] 
df.loc[ '2013-01-01':'2013-01-05', 'A':'C'] 
```





```
df.loc[ '2013-01-01', 'A'] # single value located in row '2013-01-01', col 'A'
df.loc[ '2013-01-01', ['A', 'B'] ] # 印出座標 row '2013-01-01', col 'A'和'B'的所有值
df.loc[ '2013-01-01':'2013-01-02', 'A' ] # 印出座標 row '2013-01-01'到'2013-01-02', col 'A'的所有值
df.loc[ '2013-01-01':'2013-01-05', 'A':'C'] # 印出座標 row '2013-01-01'到'2013-01-05', col 'A'到'C'的所有值
```

### 2. 請根據提供的資料，選擇出下列的要求：

```
# - select the first 3 rows.
# - select the odd rows. (index = 1, 3, 5)
# - select the last 2 columns.
# - select the even columns. (index = 0, 2, 4)
```


````python
# 記得先 Import 正確的套件
import pandas as pd
import numpy as np
````

````python
df = pd.DataFrame(np.random.randn(6,3),columns=['col1','col2','col3'])
df
````

````python
print('first 3 rows:\n', df[0:3])
print('\nodd rows:\n', df.iloc[1::2])
print('\nlast 2 columns:\n', df.iloc[:, -2:])
print('\neven columns:\n', df.iloc[:, 0::2])
````

### 3. 請根據提供的資料，選擇出下列的要求：

```
# - 1. filtered by first column > 20?
# - 2. filtered by first column + second column > 50
# - 3. filtered by first column < 30 or second column > 30
# - 4. filtered by total sum of row > 100
```

````python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
````

````python
print('first col > 20:\n', df[df[0] > 20])
print('\nfirst column + second column > 50:\n', df[df[0] + df[1] > 50])
print('\nfirst column < 30 or second column > 30:\n', df[(df[0] < 30) | (df[1] > 30)])
print('\ntotal sum of row > 100:\n', df[df.sum(axis=1) > 100])
````
