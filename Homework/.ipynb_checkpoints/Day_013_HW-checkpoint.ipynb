````python
# 記得先 Import 正確的套件
import pandas as pd
import numpy as np
````

### 1. 請接下列資料依照指定規則做合併：

```
- 依照 fruit 欄位做合併
- 依照 index 索引做合併
```


````python
df1 = pd.DataFrame({
    'fruit': ['apple', 'banana', 'orange'] * 3,
    'weight': ['high', 'medium', 'low'] * 3,
    'price': np.random.randint(0, 15, 9)
})

df2 = pd.DataFrame({
    'fruit': ['apple', 'orange', 'pine'] * 2,
    'weight': ['high', 'low'] * 3,
    'price': np.random.randint(0, 15, 6)
})
print(df1)
print(df2)
````

````python
# - 依照 fruit 欄位做合併
pd.merge(df1, df2, on = 'fruit')
````

````python
# - 依照 index 索引做合併(= 同index的合併在一起 = 左右合併)
df1.join(df2)
````

### 2. [簡答題] 承上題，請問為什麼依照 merge 合併後有些資料不見了？


````python
# 因為.merge()的parameter default how = 'inner'，只會留下df1和df2都有的key之比較
# 完整保留：指定how = 'outer'
pd.merge(df1, df2, on = 'fruit', how = 'outer')
````

### 3. [簡答題] 承上題，請問為什麼依照 index 合併會發生錯誤？請用程式解決。

因為df1和df2的column name有重複，在.join()中，若不特別說明重複col如何處理，就會產生error（不像merge會自動補_x、_y）

````python
df1.join(df2, lsuffix = '_df1', rsuffix = '_df2')
````
